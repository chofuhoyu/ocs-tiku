```python
VOLCANO_API_URL = "https://ark.cn-beijing.volces.com/api/v3/chat/completions"
VOLCANO_API_KEY = "028c285e-2a60-4349-8e22-ed178c751952"
VOLCANO_MODEL_ID = "doubao-seed-1-6-251015"
```

```python
PROMPT = """# Prompt：
请你根据我提供的题目进行作答，题目仅包含两种类型：选择题和判断题，具体格式如下：
1. 选择题格式：{'question': '题目内容', 'options': ['选项A文本', '选项B文本', '选项C文本', '选项D文本']}（选项数量可能为2个及以上，以实际提供为准）
2. 判断题格式：{'question': '题目内容', 'options': ['对', '错']}

请严格遵循以下作答要求：
- 仅输出最终答案，不包含任何思考过程、解释说明、额外文字或格式符号；
- 选择题直接输出正确选项对应的完整文本；
- 判断题直接输出'对'或'错'。
- 输出答案严格为options中提供的文本，不能有任何改动。

示例：
若题目为{'question': '在安全管理方面，政府部门和教育科研机构应当：', 'options': ['完善规范与标准，组织教育和培训，加强监督与检查等措施。', '完善规范与标准，组织教育和培训，让研究人员学会自我监督。', '提高人们的实验室安全保护意识，促进形成“安全第一”的实验室安全文化氛围，将实验次数降到最低。', '完善规范与标准，组织教育和培训，让不同的研究机构相互监督。']}，请输出：完善规范与标准，组织教育和培训，加强监督与检查等措施。
若题目为{'question': '研究团队负责人和实验室负责人应对所有实验参与者的安全承担起不可推卸责任。', 'options': ['对', '错']}，请输出：对

现在请接收题目并作答。

"""
```

```python
question = {'question': '实验室事故的原因是多种多样的，必须加强全方位的安全监管。',
                'options': ['对', '错']}
```

```python
message = PROMPT + "\n" + str(question)
```

```python
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {VOLCANO_API_KEY}"
}

payload = {
    "model": VOLCANO_MODEL_ID,
    "reasoning_effort": "minimal",
    "messages": [
        {"role": "user", "content": message}
    ],
    "max_completion_tokens": 65535
}
```

```python
import requests
response = requests.post(VOLCANO_API_URL, headers=headers, json=payload)
response.raise_for_status()
result = response.json()
```
